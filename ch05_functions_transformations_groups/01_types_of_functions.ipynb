 # 1. Types of Functions

```python
from spb import (
    plot,
    plot_list,
    plot_piecewise
)
from sympy import (
    asin,
    cos,
    E,
    Eq,
    integrate,
    log,
    pi,
    Piecewise,
    series,
    simplify,
    sin,
    solve,
    sqrt,
    Symbol,
    symbols
)
```

```python
x, y, z = symbols('x y z')

f1 = 2 + x
f2 = 3*y
f3 = z**2

plot(
    f1,
    f2.subs(y, x),
    f3.subs(z, x),
    (x, -10, 10)
)
```

```python
f4 = f3.subs(z, f2.subs(y, f1))
quadratic_function = 36 + 36*x + 9*x**2

plot(
    f4,
    quadratic_function,
    (x, -10, 10),
    rendering_kw=[{}, {"linestyle": "-."}]
)
```

```python
f5 = (3*x**2 - 1) / (x**2 - 3*x + 2)

plot(
    f5,
    (x, -8, 8),
    adaptive=False,
    is_point=True,
    only_integers=True
)
```

```python
f5.subs(x, 1)
```

```python
f5.subs(x, 2)
```

```python
f6 = sqrt(x)

plot(f6)
```

```python
f7 = Piecewise(
    (x, x >= 0),
    (-x, x < 0)
)

plot_piecewise(f7)
```

```python
def rational(x):
    if simplify(x).is_irrational:
        return 1
    else:
        return 0

x_coord = [-pi, -3, -E, -2, -pi/2, -1, 0, 1, pi/2, 2, E, 3, pi]
y_coord = [rational(coord) for coord in x_coord]

plot_list(
    x_coord,
    y_coord,
    adaptive=False,
    is_point=True
)
```

```python
theta = Symbol('')

plot(
    cos(theta),
    sin(theta),
    (theta, -pi, pi)
)
```

```python
series(sin(theta), theta)
```

```python
plot(asin(theta), (theta, -1, 1))
```

```python
i1 = (1-x**2)**-1/2

integrate(i1, x)
```

```python
plot(i1)
```

```python
plot(E**x, log(x), log(E**x), (x, -E, E))
```

Following are some examples of physical formulas based on elementary laws:

```python
p, w, V = symbols('p w V')

w = p*V

plot(
    w.subs(V, 1),
    w.subs(V, 2),
    w.subs(V, 4),
    (p, 0, 10)
)
```

```python
g, s, t = symbols('g s t')

gravity_acceleration = 9.807
s = g*t**2/2

plot(
    s.subs(g, gravity_acceleration),
    (t, 0, 10)
)
```

```python
k, P, T = symbols('k P T')

air_density = 1.225 # kg/m-cubed
pressure = 101325 # Pa
volume = 1/air_density
temperature = 288.15 # K

ideal_gas = Eq(P*V, k*T)

solve(
    ideal_gas.subs({P: pressure, V: volume, T: temperature}),
    k
)
```
