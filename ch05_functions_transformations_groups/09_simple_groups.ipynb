 # 9. Simple Groups

```python
from itertools import combinations
from sympy import init_printing
from sympy.combinatorics import (
    AlternatingGroup,
    CyclicGroup
)

init_printing(perm_cyclic = False)
```

```python
def is_simple(G):
    """ Check whether a group is a simple group. """

    if G.is_trivial:
        return False
    
    els = G._elements

    subgroups = {
        G.subgroup(list(set(c)))
        for r in range(len(els))
        for c in combinations(els, r+1)
    }

    for s in subgroups:
        if not all([e == G.identity for e in s._elements]):
            if not all([e in s._elements for e in els]):
                if s.is_normal(G):
                    return False

    return True


Z1 = CyclicGroup(1)
is_simple(Z1) == False
```

```python
Z2 = CyclicGroup(2)
is_simple(Z2) == True
```

```python
Z3 = CyclicGroup(3)
is_simple(Z3) == True
```

```python
Z4 = CyclicGroup(4)
is_simple(Z4) == False
```

```python
Z5 = CyclicGroup(5)
is_simple(Z5) == True
```

```python
A4 = AlternatingGroup(4)
is_simple(A4) == False
```
