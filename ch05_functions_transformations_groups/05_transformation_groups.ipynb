 # 5. Transformation Groups

```python
from sympy import (
    Eq,
    factorial,
    init_printing,
    Matrix,
    sign,
    Symbol
)
from sympy.combinatorics.named_groups import (
    AlternatingGroup,
    SymmetricGroup
)

init_printing(perm_cyclic = False)
```

 > All told, the initial and quite intuitive observations of symmetry, of motion, and of transformations in geometry and of their composition leads to the formal set-theoretic notion of a transformation group. Its study involves algebra, geometry, and continuity.

```python
n = 3
Sn = SymmetricGroup(n)
Sn
```

```python
Sn_perms1 = list(Sn.generate_dimino())
Sn_perms1
```

```python
Sn_perms2 = list(Sn.generate_schreier_sims())
Sn_perms2
```

```python
Eq(factorial(n), len(Sn_perms1))
```

```python
An = AlternatingGroup(n)
An
```

```python
An.is_subgroup(Sn)
```

```python
An_perms1 = list(An.generate_dimino())
An_perms1
```

```python
An_perms2 = list(An.generate_schreier_sims())
An_perms2
```

```python
Eq(factorial(n)/2, len(An_perms1))
```

```python
Sn_even_trans = [x for x in Sn_perms1 if len(x.transpositions()) % 2 == 0]
Sn_even_trans
```

```python
set(An_perms1) == set(Sn_even_trans)
```

```python
def product(x1, x2, x3, x4):
    return (x1-x2)*(x1-x3)*(x1-x4)*(x2-x3)*(x2-x4)*(x3-x4)

S4 = SymmetricGroup(4)
S4_pos = [x for x in S4.elements if sign(product(x(0), x(1), x(2), x(3))) == 1]

A4 = AlternatingGroup(4)

set(S4_pos) == set(A4.elements)
```

```python
F = Symbol('F', real=True)
n = 2

V = Matrix(n, n, lambda i, j: F)
V
```

```python
V0 = V.subs(F, 0)
V0
```

```python
V0.det()
```

```python
V0[0, 0] = 1
V0[1, 1] = 1
V0
```
