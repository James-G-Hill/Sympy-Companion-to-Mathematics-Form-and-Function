 # 2. Maps

```python
from sympy import (
    Function,
    Matrix,
    symbols
)
```

 > ... "maps" - that is, functions, from one set *X* of objects, points, or numbers into some other set *Y*.

```python
a, b, c, d, x, y = symbols('a b c d x y')

class T(Function):
    @classmethod
    def eval(cls, x, y):
        return a*x + b*y, c*x + d*y

T(1, 0)
```

```python
T(0, 1)
```

```python
t1 = T(2, 3)
t1
```

```python
t1.subs({a: 1, b: 0, c: 0, d: 1})
```

```python
m = Matrix([['a', 'b'], ['c', 'd']])
m
```

```python
mm = m*m
mm
```

```python
m_basis = m.subs({'a': 1, 'b': 0, 'c': 0, 'd': 1})
m_basis
```

```python
m_basis * m_basis
```

```python
m23 = Matrix([2, 3])
m23
```

```python
m_basis * m23
```
