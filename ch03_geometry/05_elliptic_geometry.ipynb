 ## 5. Elliptic Geometry

```python
from sympy import (
    cos,
    ln,
    pi,
    sin,
    symbols,
    tan
)

from sympy.plotting import (
    plot,
    plot_parametric,
    plot3d_parametric_surface
)
```

 The elliptic geometry is 1 in which there are no parallel lines.  The behaviour of lines in such a geometry is similar to latitude / longitude lines on a sphere.

 > The geometry of great circles on the sphere *is* a non-Euclidean geometry.

```python
def tractrix():
    """
    A curve in the xy plane.
    """
    theta = symbols('theta')
    x = sin(theta)
    y = cos(theta) + ln(tan(theta / 2))
    return x, y


tractrix()
```

```python
plot_parametric(
    tractrix(),
    (0, 1),
    adaptive=False,
    nb_of_points=1000
)
```

```python
plot(*tractrix())
```

```python
def rotate_axis_y(fnc):
    """
    Rotate a 2d function around the 'y' axis.

    fnc: A function.
    """
    theta = symbols('theta')
    u, v = symbols('u, v')
    x, y = fnc

    x = x.subs(theta, u)
    y = y.subs(theta, u)
    
    z = x * -sin(v)
    x = x * cos(v)

    return (x, y, z)


rotate_axis_y(tractrix())
```

The function below creates a pseudosphere in this case.
It is also known as a tractricoid.

```python
def plot_rotated(fnc):
    """
    Plot a rotated function.

    fnc: A rotated function.
    """
    u, v = symbols('u, v')
    plot3d_parametric_surface(
        (*fnc, (u, 0, 1), (v, -pi, pi)),
        nb_of_points_u=1000,
        nb_of_points_y=1000
    )


plot_rotated(rotate_axis_y(tractrix()))
```
