 ## 10. Geometry by Groups

```python
from geometry.rigidMotion import (
    GroupEuclidean,
    RigidMotion
)
from geometry.plane import Plane
from geometry.plot import plane_plot
from geometry.reflection import Reflection
from geometry.rotation import Rotation
from sympy import (
    Line,
    pi,
    Point,
    Segment,
    Triangle
)
```

 An *involution* is any element of a group (i.e. function) that is of order 2; that is, the square of the function returns the identity (such as 2 reflections on a line, 2 180 degree rotations around a point, or multiplying by -1 twice).

```python
p1 = Point(-1, 0)
p2 = Point(1, 0)
t1 = Triangle(Point(0, 0), Point(.5, .5), Point(-.5, .5))
k = Line(p1, p2)
Pl1 = Plane(p1, p2, k, t1)
plane_plot(Pl1)
```

```python
Lk = Reflection(k)
Pl2 = Lk.act(Pl1)
plane_plot(Pl1, Pl2)
```

```python
Pl3 = (Lk + Lk).act(Pl1)
plane_plot(Pl3)
```

```python
RA = Rotation(pi)
Pl4 = RA.act(Pl1)
plane_plot(Pl1, Pl4)
```

```python
Pl5 = (RA + RA).act(Pl1)
plane_plot(Pl5)
```

 **Theorem 1**
 A rigid motion M of the plane which is an involution is either a reflection $L_k$ in some line k or a half-turn $R_A$ about some point A.

```python
B = Point(0, 1)
l = k
Pl6 = Plane(p1, p2, l)
Pl7 = Plane(B)
M = RA
C = M.act(Pl7)
BC = Plane(Segment(B, C[0]))
plane_plot(Pl6, Pl7, C, BC)
```

```python
M2B = M.act(M.act(Plane(B)))[0]
B.equals(M2B)
```

```python
Pl8 = M.act(Pl6)
Pl9 = Plane(Point(0, 0))
plane_plot(Pl6, Pl8, Pl9)
```

 **Theorem 2**
 Every rigid motion is a composite of involutions.

```python
k2 = Line(Point(0, 0), Point(1, 1))
l2 = Line(Point(0, 0), Point(1, -1))
p0 = k2.intersection(l2)
t2 = Triangle(Point(0, 0), Point(.5, .5), Point(0, .5))
Pl10 = Plane(k2, l2, *p0, t2)
plane_plot(Pl10)
```

```python
Lk1 = Reflection(k2)
Ll1 = Reflection(l2)
Pl11 = Ll1.act(Pl10)
Pl12 = Lk1.act(Pl11)
plane_plot(Pl10, Pl11, Pl12)
```

```python
k3 = Line(Point(0, 0), Point(1, 0))
l3 = Line(Point(0, 1), Point(1, 1))
m = k3.perpendicular_line(Point(0, 0))
Lk2 = Reflection(k3)
Ll2 = Reflection(l3)
Pl13 = Plane(k3, l3, m, t2)
Pl14 = Lk2.act(Pl13)
Pl15 = Ll2.act(Pl14)
plane_plot(Pl13, Pl14, Pl15)
```

```python
p3 = Point(0, 0)
RA2 = Rotation(pi, p3)
k4 = Line(Point(-1, 0), Point(1, 0))
Lk3 = Reflection(k4)
Pl16 = Plane(p3, k4, t2)
Pl17 = Lk3.act(RA2.act(Pl16))
plane_plot(Pl16, Pl17)
```

```python
Pl18 = RA2.act(Lk3.act(Pl16))
plane_plot(Pl16, Pl18)
```
