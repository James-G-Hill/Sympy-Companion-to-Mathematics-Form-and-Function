```python
hk_split_point = t.altitudes[Point(0,0)].args[1]
hk_split_point
```

```python
h = Segment2D(hk_split_point, t.args[2])
h
```

```python
h.length
```

```python
k = Segment2D(t.args[1], hk_split_point)
k
```
