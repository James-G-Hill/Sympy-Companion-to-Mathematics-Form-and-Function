# 3. Natural Numbers Described by Recursion

```python
from sympy import (
    ConditionSet,
    Function,
    symbols,
    S
)
from sympy.categories import (
    Diagram,
    DiagramGrid,
    NamedMorphism,
    Object,
    XypicDiagramDrawer
)
```

General Point: the axioms needed to describe a Mathematical structure are themselves by no means unique. The recursion theorum for example is a convenient form of axiom.

```python
N = S.Naturals
S_ob = Object("S")

s = NamedMorphism(N, N, "s")
sDash = NamedMorphism(S_ob, S_ob, "sDash")
h = NamedMorphism(N, S_ob, "h")
i = NamedMorphism(S_ob, N, "i")

d = Diagram([s, sDash, h, i])

drawer = XypicDiagramDrawer()
print(drawer.draw(d, DiagramGrid(d)))
# https://quicklatex.com/
# Add `\usepackage[all]{xy}` to the preamble
```

```python
x = symbols("x")

N = S.Naturals0
S_ob = ConditionSet(x, N.contains(x))
S_ob
```

```python
-1 in S_ob
```

```python
0 in S_ob
```

```python
class i(Function):
    @classmethod
    def eval(cls, arg1):
        if arg1 == S_ob:
            return N
        
i(S_ob)
```

```python
class zero(Function):
    @classmethod
    def eval(cls):
        return S_ob
    
zero()
```

```python
class sDash(Function):
    @classmethod
    def eval(cls, arg1):
        if arg1 == S_ob:
            return S_ob
    
sDash(S_ob)
```

```python
class h(Function):
    @classmethod
    def eval(cls, arg1):
        if arg1 == N:
            return S_ob
        
h(N)
```

```python
class f(Function):
    @classmethod
    def eval(cls, arg1):
        if arg1 == N:
            return i(h(arg1))
        
f(N)
```

```python
class s(Function):
    @classmethod
    def eval(cls, arg1):
        if arg1 == N:
            return N
        
s(N)
```

```python
sDash(h(N)) == h(s(N))
```

```python
s(f(N)) == f(s(N)) # identity function
```

```python
f(N) == i(h(N))
```
