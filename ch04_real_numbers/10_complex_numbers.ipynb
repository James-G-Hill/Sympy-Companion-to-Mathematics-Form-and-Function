 ## 10. Complex Numbers

```python
from spb import (
    plot_complex_list,
    plot_geometry
)
from sympy import (
    cos,
    Eq,
    I,
    pi,
    Point,
    Segment,
    sin,
    solveset,
    Symbol,
    symbols
)
```

 The complex numbers **C** arise in order to solve plausible polynomial equations such as $x^2 + 1 = 0$ and $x^2 + x + 1 = 0$.
 In this case, we see that $x^2$ must be equal to -1, but $x$ must be equal to 0.

```python
a, b, c, d = symbols('a b c d', real=True)
c1, c2, c3, c4 = symbols('c1 c2 c3 c4')

lhs_add = (a + b*I) + (c + d*I)
rhs_add = (a + c) + (b + d)*I
com_add = Eq(lhs_add, rhs_add)
com_add
```

```python
c1 = lhs_add.subs({a: 2, b: 3, c: 4, d: 5})
c1
```

```python
rhs_add.subs({a: 2, b: 3, c: 4, d: 5})
```

```python
com_add.subs({a: 2, b: 3, c: 4, d: 5})
```

```python
lhs_mul = (a + b*I)*(c + d*I)
rhs_mul = (a*c - b*d) + (a*d + b*c)*I
com_mul = Eq(lhs_mul, rhs_mul)
com_mul
```

```python
c2 = lhs_mul.subs({a: 2, b: 3, c: 4, d: 5}).n()
c2
```

```python
rhs_mul.subs({a: 2, b: 3, c: 4, d: 5})
```

```python
com_mul.subs({a: 2, b: 3, c: 4, d: 5}).n()
```

```python
def ari_add(ab, cd):
    a, b = ab
    c, d = cd
    return a + c, b + d

ari_add((2, 3), (4, 5))
```

```python
def ari_mul(ab, cd):
    a, b = ab
    c, d = cd
    return a*c - b*d, a*d + b*c

ari_mul((2, 3), (4, 5))
```

```python
ari_i = 0, 1

ari_mul(ari_i, ari_i)
```

```python
ari_add((2, 0), (0, 3))
```

```python
p0 = Point(0, 0)
z1 = Point(*c1.as_real_imag())
z2 = Point(*c2.as_real_imag())

plot_geometry(
    (p0, '0'),
    (Segment(p0, z1), 'Z1'),
    (Segment(p0, z2), 'Z2')
)
```

```python
r, theta = symbols('r θ')

polar = r*(cos(theta) + I*sin(theta))
polar.subs({r: 1, theta: 0})
```

```python
polar.subs({r: 1, theta: pi/2})
```

```python
plot_complex_list(
    ([polar.subs({r: 1, theta: y}).n() for y in range(6)], 'orange'),
    ([polar.subs({r: 2, theta: y}).n() for y in range(6)], 'blue'),
    ([polar.subs({r: y/2.5, theta: y/2.5}).n() for y in range(10)], 'green'),
    aspect='equal'
)
```

```python
s, phi = symbols('s φ')

polar_add = r*s*(cos(theta + phi) + I*sin(theta + phi))
polar_add
```

```python
polar_add.subs({r: 2, theta: 0, s: 3, phi: pi/2}).n()
```
