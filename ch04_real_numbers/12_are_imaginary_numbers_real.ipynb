 ## 12. Are Imaginary Numbers Real?

```python
from spb import (
    plot,
    plotgrid,
    plot_complex,
    plot_list
)
from sympy import (
    E,
    I,
    pi,
    rootof,
    sin,
    solve,
    solveset,
    sqrt,
    Symbol,
    symbols
)
```

 > If Mathematics is the science of number, space, time, and motion, and if number covers just whole, rational and real numbers, then there is no place in this science for these imaginary numbers. Therefore, the gradual general acceptance (after 1800) of the complex numbers as reasonable and useful does mark a departure from the "number, space, time, and motion" concept of Mathematics.

```python
i, x = symbols('i x')

x_limit = 9
x_range = range(x_limit+1)

uni_polys = [x**2-i for i in x_range]
root_pos = [rootof(i, 1).n() for i in uni_polys]
root_neg = [rootof(i, 0).n() for i in uni_polys]

p1 = plot(x, (x, 0, x_limit), show=False, aspect='equal')
p2 = plot_list(
    (x_range, root_pos, 'positive roots'),
    (x_range, root_neg, 'negative roots'),
    show=False
)

(p1 + p2).show()
```

```python
p3 = plot(E**x, (x, -pi, pi), show=False)

p4 = plot_complex(
    E**x,
    (x, -pi+I*-pi, pi+I*pi),
    threed=True,
    use_cm=True,
    show=False
)

fig1 = plotgrid(p3, p4, nr=1, nc=-1)
```

```python
p5 = plot(sin(x), (x, -pi, pi), show=False)

p6 = plot_complex(
    sin(x),
    (x, -pi+I*-pi, pi+I*pi),
    threed=True,
    use_cm=True,
    show=False
)

fig2 = plotgrid(p5, p6, nr=1, nc=-1)
```
