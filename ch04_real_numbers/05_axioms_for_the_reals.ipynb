 ## 5. Axioms for the Reals

```python
from sympy import (
    And,
    ConditionSet,
    Contains,
    Implies,
    Interval,
    Naturals,
    oo,
    Symbol,
    symbols
)
```

```python
a = Symbol('a', negative=False, real=True)
b = Symbol('b', negative=False, real=True)
n = symbols('n')

S = ConditionSet(n, n*a <= b, Naturals)
S
```

```python
b_star = Symbol('b*', negative=False, real=True)
b_star
```

```python
b_star >= n*a
```

```python
b_star >= (n+1)*a
```

```python
b_star - a >= n*a
```

```python
b_star - a < b_star
```

 There are multiple forms of the completeness axiom, demonstrating that the field of **R** is complete & has no gaps:

 - Completeness Axiom
 - Dedekind Cut Axiom
 - Caunch Condition
 - Weierstrass Condition

 The Completeness Axiom is defined as:

 > Every non-empty set of reals with an upper bound has a least upper bound

```python
x, y = symbols('x, y', real=True)
r = Symbol('r', real=True)

L = Interval(-oo, y, right_open=True)
U = Interval(x, oo, left_open=True)

Implies(
    Implies(And(Contains(x, L), Contains(y, U)), x < y),
    And(x <= r, r <= y)
)
```
