 ## 6. Arithmetic Construction of the Reals

```python
from spb import plot
from sympy import (
    And,
    ConditionSet,
    Contains,
    Implies,
    oo,
    Rationals,
    SeqAdd,
    sequence,
    Sum,
    Symbol,
    symbols,
    Union
)
```

 There are different possible constructions of the reals through the use of arithmetic & sets

 ### Dedekind Cut

 This method defines a real number as a lower set divided by a cut in the field of rational numbers **Q**.
 A real number is then defined as the lower set.

```python
x, y = symbols('x, y', rational=True)

L = ConditionSet(x, x < y, base_set=Rationals)
U = ConditionSet(y, x < y, base_set=Rationals)

dedekind_cut = Union(L, U)
dedekind_cut
```

```python
x_p = Symbol('x`')

Implies(x_p < x, Contains(x_p, L))
```

 ### Cauchy Sequences

 This method defines equivalence classes of Cauchy sequences that converge to the same limit as being real numbers.

```python
n = Symbol('n')

seq1 = sequence(1-(1/n), (1, oo))
seq1
```

```python
seq2 = sequence(n/(n+2), (1, oo))
seq2
```

```python
seq3 = SeqAdd(seq1, -seq2)
seq3
```

```python
eulers = sequence((1+(1/n))**n, (1, oo))
eulers
```

```python
eulers.formula
```

```python
plot(
    seq1.formula,
    seq2.formula,
    seq3.formula,
    eulers.formula,
    (n, 1, 100)
)
```

```python
c = Sum(1/(n*(n+1)), (n, 1, oo))
c
```

```python
c.doit()
```

```python
c_p = Sum(((n+1)/(n+3))-(n/(n+2)), (n, 0, oo))
c_p
```

```python
c_p.doit()
```

```python
And(
    c.doit() <= c_p.doit(),
    c_p.doit() <= c.doit()
)
```

```python
c_add_c_p = (c + c_p)
c_add_c_p
```

```python
c_add_c_p.doit()
```

```python
c_prod_c_p = (c * c_p)
c_prod_c_p
```

```python
c_prod_c_p.doit()
```

```python
p = Symbol('p')

plot(
    c.subs(oo, p),
    c_p.subs(oo, p),
    c_add_c_p.subs(oo, p),
    c_prod_c_p.subs(oo, p),
    (p, 1, 50),
    adaptive=False,
    only_integers=True,
    is_point=True,
    is_filled=True
)
```
