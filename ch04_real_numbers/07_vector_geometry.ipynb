 ## 7. Vector Geometry

```python
from spb import plot_geometry
from sympy import (
    Eq,
    Point,
    Segment,
    Symbol,
    symbols
)
from sympy.vector import CoordSys3D
```

```python
p0 = Point(0, 0)
pU = Point(1, 0)
unit = Segment(p0, pU).length
unit
```

```python
pP = Point(7, -2)
p0pP = Segment(p0, pP)

pQ = Point(4, 3)
p0pQ = Segment(p0, pQ)

pR = pP + pQ
p0pR = Segment(p0, pR)

plot_geometry(
    (p0, '0'),
    (pP, 'P'),
    (pQ, 'Q'),
    (pR, 'R'),
    (p0pP, '0P'),
    (p0pQ, '0Q'),
    (p0pR, '0R')
)
```

```python
def vectorMultiplication(scalar, vector):
    return scalar * vector.p2

r = Symbol('r', real=True)
v = p0pR
rv = vectorMultiplication(r, v)
rv
```

```python
plot_geometry(
    (p0, '0'),
    (pR, 'R'),
    (Segment(p0, rv.subs(r, 3)), '+rv'),
    (Segment(p0, rv.subs(r, -3)), '-rv'),
    aspect='equal',
    ylim=(-10, 10)
)
```

```python
def weighted_average(*args):
    w, p = zip(*args)
    if sum(w) == 1:
        P = Point(0, 0)
        for x, y in zip(w, p):
            P += (x*y)
        return P
    else:
        raise ValueError('Weights don\'t add up to 1')

weighted_average(
    (.5, Point(3, 8)),
    (.5, Point(-3, 4))
)
```

```python
weighted_average(
    (.25, Point(-6, 3)),
    (.25, Point(1, 5)),
    (.25, Point(-6, -6)),
    (.25, Point(3, -1))
)
```
