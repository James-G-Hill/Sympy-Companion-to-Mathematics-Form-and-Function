*Transformations*:

- a permutation of a set

```python
perm_1 = Permutation([1, 0, 3, 2])
perm_2 = Permutation([3, 2, 1, 0])
perm_3 = Permutation([2, 3, 0, 1])

four_group = PermutationGroup(perm_1, perm_2, perm_3)
four_group
```

```python
class vertex():
    def __init__(self, ident, coord):
        self.ident = ident
        self.coord = coord
    
verts = [
    vertex('a', (0, 0)),
    vertex('b', (1, 0)),
    vertex('c', (1, 1)),
    vertex('d', (0, 1))
]

def f(verts):
    """ Assigns a label to each vertice in X. """
    labels = {'a': 1, 'b': 2, 'c': 3, 'd': 4}
    return [(labels[vert.ident], vert.coord) for vert in verts]

f(verts)
```

```python
def T(X, perm):
    """ Transforms the vertices in X by the given permutation. """
    return perm(X)

labelled = f(verts)
T_f = T(labelled, four_group[0])
T_f
```

```python
Y = T(verts, four_group[0])
f_hashT = f(Y)
f_hashT
```

```python
T_f == f_hashT
```

```python
def f_inverse(labelled_verts):
    return [vert[1] for vert in labelled_verts]

f_inverse(T_f)
```

```python
hashT_equivalent = f_inverse(T(f(verts), four_group[0]))
hashT_equivalent
```

```python
[v.coord for v in T(verts, four_group[0])] == hashT_equivalent
```

The above shows a generalization of the fact that permutations of a typical set represents the permutations of any other set of equal size; the process is related to the bijection of a set to cardinal numbers and then a comparison against another similarly sized set.
