A group *G* has 3 rules:

- element *st* from *r*, *s*, *t* in *G* is an associative product
- element *e* of *G* is such that *te* = *t*
- element *t* has an inverse such that *t* x *tInverse* = *e*

```python
G = SymmetricGroup(3)
G
```

```python
T = G._elements
T
```

```python
r = T[1]
s = T[3]
t = T[5]

# The product of the group is associative
r * (s * t) == (r * s) * t
```

```python
e = G.identity
e
```

```python
t
```

```python
t * e
```

```python
# The identity element (unit) in composition returns whatever composes against it
t * e == t
```

```python
s * e == s
```

```python
r
```

```python
r ** -1
```

```python
~r # shortcut for the inverse
```

```python
r * (r ** -1)
```

```python
# Each permutation has an inverse
r * (r ** -1) == e
```

```python
t * e == t == e * t
```

```python
t * ~t == e == ~t * t
```

```python
G = SymmetricGroup(3)
H = DihedralGroup(3)

w, x = G.generators
y, z = H.generators

b = homomorphism(G, H, [w, x], [y, z], check = False)
b.kernel()
```

```python
b.is_injective()
```

```python
b.is_isomorphism()
```

```python
b.image()
```

```python
b.domain
```

```python
b.codomain
```

```python
b.images
```

```python
S = b.domain._elements[1]
S
```

```python
T = b.domain._elements[3]
T
```

```python
S * T
```

```python
b(S)
```

```python
b(T)
```

```python
b(S * T)
```

```python
b.invert(b(S * T))
```

```python
S * T == b.invert(b(S * T)) ==  b.invert(b(S)) *  b.invert(b(T))
```

```python
b(S * T) == b(S) * b(T)
```

```python
b(b.domain.identity) == b.domain.identity
```

```python
b(~S) == ~b(S)
```

> But geometry is not the only source of the idea of isomorphism.

```python
x = 5
log_x = log(x, 10)
log_x.evalf()
```

```python
y = 11
log_y = log(y, 10)
log_y.evalf()
```

```python
xy = x * y
xy
```

```python
log(xy, 10).evalf()
```

```python
(log(x, 10) + log(y, 10)).evalf()
```

```python
log(xy, 10).evalf() == (log(x, 10) + log(y, 10)).evalf()
```

```python
log(1, 10).evalf()
```

```python
log(x**-1, 10).evalf()
```

```python
-log(x, 10).evalf()
```

```python
log(x**-1, 10).evalf() == -log(x, 10).evalf()
```
