 # 9. The Principal Function

 > A conservative mechanical system with Legrangian $L$ determines a "principal function" $S$ as the integral of $L$ along the trajectories. As a function of time and the end-point of the trajectory, $S$ satisfies the Hamilton-Jacobi (HJ) partial differential equation. Conversely ... suitable solutions of the HJPDE give the trajectories, often called the *characteristics* of the P.D.E.
