 # 3. Eigenvalues

```python
from sympy import (
    Derivative,
    Eq,
    eye,
    Idx,
    IndexedBase,
    Matrix,
    MatrixSymbol,
    Sum,
    Symbol,
    symbols,
    Tuple,
    zeros
)
```

```python
i = Idx('i')
j = Idx('j')

m, n = symbols('m, n')

x = IndexedBase('x')
y = IndexedBase('y')

P = MatrixSymbol('P', 3, 3)

eq1 = Eq(y[i], Sum(P[i, j]*x[j], (i, m, n)))
eq1
```

```python
X = Matrix([[1, 2, 3]])
Y = Matrix([[2, 1, 3]])

P_subs = Matrix([
    [0, 1, 0],
    [1, 0, 0],
    [0, 0, 1]
])

eq1_subs = {
    x: Tuple(*X),
    y: Tuple(*Y),
    m: 0,
    n: len(Tuple(1, 2, 3))-1,
    P: P_subs
}

[
    eq1.subs(eq1_subs).subs({j: Tuple(1, 0, 2)[e], i: e}).doit()
    for e in range(3)
]
```

```python
A = Matrix([
    [1, 0, 0],
    [0, 0, 1],
    [0, 1, 0]
])

A
```

```python
X
```

```python
X_p = X*A
X_p
```

```python
B = Matrix([
    [0, 0, 1],
    [0, 1, 0],
    [1, 0, 0]
])

B
```

```python
Y
```

```python
Y_p = Y*B
Y_p
```

```python
Eq(Y, X*P_subs)
```

```python
X_p*P_subs
```

```python
Eq(Y_p, X_p*P_subs)
```

```python
Eq(B, P_subs.inv()*A*P_subs)
```

```python
A.eigenvals(multiple=True)
```

```python
lambda_eigen = list(set(A.eigenvals(multiple=True)))
lambda_eigen
```

```python
X_eigen_vecs = A.eigenvects()
X_eigen_vecs
```

```python
lamb = X_eigen_vecs[0][0]
X_eigen = X_eigen_vecs[0][2][0]

Eq(A * X_eigen, X_eigen * lamb)
```

```python
all([
    Eq(X_eigen_vecs[e][0]*X_eigen_vecs[e][2][0], A*X_eigen_vecs[e][2][0])
    for e in range(len(X_eigen_vecs))
])
```

```python
id_matrix = eye(3)
id_matrix
```

```python
Eq(
    (A - lamb*id_matrix) * X_eigen,
    zeros(3, 1)
)
```

```python
def is_singular(M):
    return M.det() == 0

is_singular((A - lamb*id_matrix))
```

```python
A_poly = A.charpoly()
A_poly
```

```python
all([A_poly.eval(m) == 0 for m in lambda_eigen])
```

```python
A_poly.all_roots()
```

```python
list(set(A_poly.all_roots())) == lambda_eigen
```
