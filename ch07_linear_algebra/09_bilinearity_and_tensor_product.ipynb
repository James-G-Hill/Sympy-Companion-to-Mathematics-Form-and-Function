 # 9. Bilinearity and Tensor Product

```python
from spb.backends.plotly import PB
from spb.defaults import cfg
from spb.plot_functions import plot3d
from sympy import (
    Eq,
    Idx,
    Lambda,
    Matrix,
    MatrixSymbol,
    Sum,
    Symbol,
    symbols
)
from sympy.physics.quantum import TensorProduct

cfg['matplotlib']['show_minor_grid'] = True
```

```python
U = MatrixSymbol('U', 2, 1)
V = MatrixSymbol('V', 2, 1)

B = Lambda((U, V), U.T * V)
B(U, V)
```

```python
W = B(U, V)
W
```

```python
Eq(B(U, V), W)
```

```python
u = Matrix([1, 2])
u
```

```python
v = Matrix([3, 4])
v
```

```python
Buv = B(u, v).doit()[0]
Buv
```

```python
Eq(Buv, u.dot(v))
```

```python
v_fixed = (Matrix([1, 3]), Matrix([2, 4]), Matrix([3, 5]))

x, y = symbols('x, y')

plot3d(
    B(Matrix([x, y]), v_fixed[0]).doit()[0],
    B(Matrix([x, y]), v_fixed[1]).doit()[0],
    B(Matrix([x, y]), v_fixed[2]).doit()[0],
    (x, -10, 10),
    (y, -10, 10),
    backend=PB,
    camera={'eye': {'x': -2, 'y': -2, 'z': 2}},
    rendering_kw={'opacity': .8}
)
```

```python
UV = MatrixSymbol('U⊗V', 2, 2)
w = Symbol('W')
T = Lambda(UV, w)
T
```

```python
lin_inn_prod = Lambda(UV, UV[0, 0] + UV[1, 1])

T.subs(w, lin_inn_prod(UV))(UV_tp(u.T, v))
```

```python
Eq(
    B(u, v).doit()[0],
    T.subs(w, lin_inn_prod(UV))(UV_tp(u.T, v))
)
```

```python
m = 2
n = 2

U_mat = MatrixSymbol('U', m, 1)
V_mat = MatrixSymbol('V', n, 1)

mn = [TensorProduct(i, j) for i in U_mat for j in V_mat]
Fmn = Matrix(mn)
Fmn
```

```python
a = MatrixSymbol('a', m, n)

u_sym = MatrixSymbol('u', m, 1)
v_sym = MatrixSymbol('v', 1, n)

i = Idx('i')
j = Idx('j')

formal_linear_combination = Sum(
    a[i, j] * UV_tp(u_sym[i, 0], v_sym[j, 0]),
    (i, 0, m-1),
    (j, 0, n-1)
)

formal_linear_combination
```

```python
a_rep = [Matrix(2, 2, [x + 1, 0, 0, 1]) for x in range(3)]
a_rep[0]
```

```python
formal_linear_combination_examples = [formal_linear_combination.subs({a: x, u_sym: u, v_sym: v}) for x in a_rep]
formal_linear_combination_examples[0]
```

```python
formal_linear_combo_results = [x.doit().args[0] for x in formal_linear_combination_examples]
formal_linear_combo_results
```

```python
formal_linear_combo_results[0]
```

```python
B_right = Sum(
    a[i, j] * B(u_sym, v_sym.T),
    (i, 0, m-1),
    (j, 0, n-1)
)

B_right
```

```python
B_right_examples = [B_right.subs({a: x, u_sym: u, v_sym: v.T}) for x in a_rep]
B_right_examples[0]
```

```python
B_right_results = [x.doit()[0] for x in B_right_examples]
B_right_results
```

```python
B_right_results[0]
```

```python
formal_combo_B_right_diffs = Matrix([y - x for x, y in zip(formal_linear_combo_results, B_right_results)])
formal_combo_B_right_diffs
```

```python
z = Symbol('z')
T_symbolic = Symbol('T')
T_example = Lambda(z, T_symbolic * z)

formal_linear_combination_equality = Eq(
    T_example(formal_linear_combination),
    B_right
)

formal_linear_combination_equality
```
