```python
from sympy import (
    Eq,
    eye,
    Function,
    Idx,
    IndexedBase,
    Lambda,
    Matrix,
    MatrixSymbol,
    Ne,
    Piecewise,
    sin,
    Sum,
    Symbol,
    Tuple
)
```

```python
row_v = MatrixSymbol('r', 1, 2)
col_v = MatrixSymbol('c', 2, 1)

row_col_prod = (row_v * col_v)
row_col_prod.shape
```

```python
row_v_sub = Matrix([[2, 4]])
row_v_sub.normalized()
```

```python
col_v_sub.normalized().norm()
```

```python
row_col_prod.subs({row_v: row_v_sub, col_v: col_v_sub}).doit()
```

```python
f = Function('f')
n = Symbol('n')
i = Idx('i')
u = IndexedBase('u')
x = IndexedBase('x')

eq1 = Eq(f(Sum(u[i] * x[i], (i, 1, n))), Sum(f(u[i]) * x[i], (i, 1, n)))
eq1
```

```python
eq1_subs = {
    n: 1,
    u: Tuple(*row_v_sub),
    x: Tuple(*col_v_sub)
}

eq1.replace(f, Lambda((x), x*3)).subs(1, 0).subs(eq1_subs).doit()
```

```python
j = Idx('j')

special_vectors = (
    Piecewise(
        (Eq(u[i]*u[j], 1), Eq(i, j)),
        (Eq(u[i]*u[j], 0), Ne(i, j))
    )
)

special_vectors
```

```python
special_vectors_lamb = Lambda((u), special_vectors)

n_matrix = 3

kronecker_delta = [
    [
        int(not bool(special_vectors_lamb(Tuple(0, 0, 0)).subs({i: l, j: m})))
        for l in range(0, n_matrix)
    ]
    for m in range(0, n_matrix)
]

special_vectors_matrix = Matrix(kronecker_delta)
special_vectors_matrix
```

```python
Eq(
    special_vectors_matrix,
    eye(n_matrix)
)
```

```python
Y = Matrix([[1, 2, 3]])
X = Matrix([2, 3, 4])

Y * X
```

```python
m = Symbol('m')
n = Symbol('n')

A = MatrixSymbol('A', m, n)
A
```

```python
AX = A * X
AX
```

```python
m = 2
n = 3

A_subs = Matrix([[1, 2, 3], [4, 5, 6]])
A_subs
```

```python
AX.subs({A: A_subs})
```

```python
AX_subs = AX.subs({A: A_subs}).doit()
AX_subs
```

```python
Eq(m, AX_subs.shape[0])
```

```python
Y_2 = Matrix([[2, 3]])

YA = Y_2 * A
YA
```

```python
YA.subs({A: A_subs})
```

```python
YA_subs = YA.subs({A: A_subs}).doit()
YA_subs
```

```python
Eq(n, YA_subs.shape[1])
```

```python
V = X
W = AX_subs

def T(V):
    return A_subs * V

Eq(T(V), W)
```

```python
f_vec = Matrix([2, 3])

def f(W):
    return f_vec.dot(W)

f(W)
```

```python
def fT(f, V):
    return f(T(V))

fT(f, V)
```

```python
def T_star(W_star):
    return fT(W_star, V)

V_star = T_star(f)
Eq(V_star, fT(f, V))
```

```python
def T_p(V):
    return Matrix([[2, 3, 4], [5, 6, 7]]) * V

T_p(V)
```

```python
def T_plus_T_p(V=None):
    mat = A_subs + Matrix([[2, 3, 4], [5, 6, 7]])
    if V is None:
        return mat
    else:
        return mat * V

T_plus_T_p()
```

```python
T_plus_T_p(V)
```

```python
Eq(
    T_plus_T_p(V),
    T(V) + T_p(V)
)
```
