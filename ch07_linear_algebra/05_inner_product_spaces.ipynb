 # 5. Inner Product Spaces

```python
from sympy import (
    Abs,
    acos,
    cos,
    E,
    Eq,
    Function,
    Idx,
    IndexedBase,
    Integral,
    Le,
    Matrix,
    MatrixSymbol,
    oo,
    sin,
    Sum,
    Symbol,
    Tuple
)
```

```python
u_subs = [1, 2, 3]
y_subs = [-3, -2, -1]

u = Matrix(u_subs)
y = Matrix(y_subs)

u.dot(y)
```

```python
x_idx = IndexedBase('x')
y_idx = IndexedBase('y')

n = Idx('n')

inner_prod = x_idx[1]*y_idx[1] + x_idx[2]*y_idx[2] + x_idx[n]*y_idx[n]
inner_prod
```

```python
inner_prod_subs = {
    'x': Tuple(*u_subs),
    'y': Tuple(*y_subs),
    'n': 2
}

Eq(
    inner_prod.replace(1, 0).replace(2, 1).subs(inner_prod_subs),
    Matrix(Tuple(*u_subs)).dot(Matrix(Tuple(*y_subs)))
)
```

```python
k = Symbol('k')

x_seq = Sum((1/k)**2, (k, 1, oo))
x_seq
```

```python
x_seq.doit()
```

```python
x_seq.n()
```

```python
x_seq.is_convergent()
```

```python
y_seq = Sum(E**-k, (k, 1, oo))
y_seq
```

```python
y_seq.doit()
```

```python
y_seq.n()
```

```python
y_seq.is_convergent()
```

```python
Sum(x_seq.function + y_seq.function, (k, 1, oo)).is_convergent()
```

```python
f = Function('f')
g = Function('g')

x = Symbol('x')

lebesgue_integral = Integral(f(x) * g(x), (x, 0, 1))
lebesgue_integral
```

```python
f_subs = cos
g_subs = sin

lebesgue_integral.replace(f, f_subs).replace(g, g_subs).doit()
```

```python
theta = Symbol('θ')
v = Matrix(y_subs)

eq1 = Eq(u.dot(v), u.norm() * v.norm() * cos(theta))
eq1
```

```python
theta_sub = acos(u.dot(v) / (u.norm() * v.norm()))
theta_sub
```

```python
theta_sub.n()
```

```python
eq1.subs(theta, theta_sub)
```

```python
Le(Abs(u.dot(v)), u.norm() * v.norm())
```

```python
distance_between_end_points = (u-v).norm()
distance_between_end_points
```

```python
distance_between_end_points.n()
```

```python
Eq(
    distance_between_end_points.n(),
    sum([(s-t)**2 for s, t in zip(u, v)])**(1/2)
)
```

```python
u_i_mat = Matrix([1, 0])
u_j_mat = Matrix([0, 1])

g_ij = u_j_mat.dot(u_i_mat)
g_ij
```

```python
x_i_mat = Matrix([2, 2])
y_j_mat = Matrix([1, 3])

eq2_l = u_i_mat.dot(x_i_mat) * u_j_mat.dot(y_j_mat)
eq2_l
```

```python
i = Symbol('i')
j = Symbol('j')

x_i = IndexedBase('x')
y_i = IndexedBase('y')

g = MatrixSymbol('g', 1, 1)

sum1 = Sum(g[i, j] * x_i[i] * y_j[j], (i, 0, 1), (j, 0, 1))
sum1
```

```python
G = u_i_mat * u_j_mat.T
G
```

```python
sum_subs = {
    g: G,
    x_i: Tuple(*x_i_mat.T.tolist()[0]),
    y_j: Tuple(*y_j_mat.T.tolist()[0])
}

eq2_r = sum1.subs(sum_subs).doit()
eq2_r
```

```python
Eq(eq2_l, eq2_r)
```
