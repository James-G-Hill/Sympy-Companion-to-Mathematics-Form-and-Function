```python
from spb.backends.plotly import PB
from spb.defaults import cfg
from spb.plot_functions import plot3d_list
from sympy import (
    Add,
    Eq,
    Expr,
    Matrix,
    MatrixSymbol,
    symbols,
    Tuple
)
from sympy.physics.quantum import TensorProduct

cfg['matplotlib']['show_minor_grid'] = True
```

```python
x, y, z = symbols('x y z')

V = Matrix([x, y, z])
V
```

```python
S = Matrix([0, 0, 1])
S
```

```python
P = Matrix([[1, 0, 0],[0, 1, 0]])
P
```

```python
V_div_S = P * V # Quotient space
V_div_S
```

```python
V_ex = [
    V.subs({x: a, y: b, z: (a+b)/2})
    for a in range(-10, 11, 2)
    for b in range(-10, 11, 2)
]

V_div_S_ex = [
    P * V.subs({x: a, y: b, z: (a+b)/2})
    for a in range(-10, 11, 2)
    for b in range(-10, 11, 2)
]

plot3d_list(
    (*zip(*V_ex), 'V'),
    (*zip(*V_div_S_ex), len(V_div_S_ex) * [0], 'V_div_S'),
    scatter=True,
    backend=PB,
    camera={'eye': {'x': -2, 'y': -2, 'z': 2}},
    rendering_kw={'opacity': .8, 'marker': {'size': 5}}
)
```

```python
u_1 = Matrix([1, 0])
u_2 = Matrix([0, 1])

u = [u_1, u_2]

v_1 = Matrix([1, 0, 0])
v_2 = Matrix([0, 1, 0])
v_3 = Matrix([0, 0, 1])

v = [v_1, v_2, v_3]

class BasisElement(Expr):
    def __new__(cls, u, v):
        obj = Expr.__new__(cls)
        obj.u_index = u[0] + 1
        obj.u = u[1]
        obj.v_index = v[0] + 1
        obj.v = v[1]
        return obj
    
    def _latex(self, printer):
        return f"(u{self.u_index} \\cdot v{self.v_index})"

L_basis = Tuple(*[
    BasisElement(x, y)
    for x in enumerate(u)
    for y in enumerate(v)
])

L_basis
```

```python
a1, a2, a3, a4, a5, a6 = symbols('a1, a2, a3, a4, a5, a6')
a = [a1, a2, a3, a4, a5, a6]

L_element = Add(*[x * y for x, y in zip(a, L_basis)])
L_element
```
