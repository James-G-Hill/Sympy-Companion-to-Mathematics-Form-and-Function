 # 6. Differential Equations

```python
from spb import plot3d
from sympy import (
    Abs,
    And,
    Derivative,
    Eq,
    Function,
    Integral,
    Lambda,
    sqrt,
    symbols,
    Symbol
)
```

```python
t, x = symbols('t, x')
h = Function('h')

eq1 = Eq(Derivative(x, t), h(x, t))
eq1
```

```python
m, n = symbols('m, n')

eq1_subs = {
    x: t**2,
    h: Lambda((m, n), (m/n)*2)
}

eq1.subs(eq1_subs).subs(t, 2).doit()
```

```python
a, b = symbols('a, b')
f = Function('f')
F = Function('F')

eq2 = Eq(F(b), F(a) + Integral(f(t), (t, a, b)))
eq2
```

```python
eq2_subs = {
    a: 2,
    b: 5,
    F(a): a**2,
    F(b): b**2,
    f(t): Derivative(t**2, t)
}

eq2.subs(eq2_subs)
```

```python
eq2.subs(eq2_subs).doit()
```

```python
g = Function('g')(t)
g_p = Lambda((t), Derivative(g, t))

Eq(
    g_p(t),
    h(g, t)
)
```

```python
a, x0, t0 = symbols('a, x0, t0')

And(
    Abs(x - x0) < a,
    Abs(t - t0) < a
)
```

```python
plot3d(
    (h(x, t)).subs(h, Lambda((m, n), 2*m/n)),
    (t, t0.subs(t0, 0), a.subs(a, 5)),
    (x, x0.subs(x0, 0), a.subs(a, 5)),
    zlim=(0, 10)
)
```

```python
delta = Symbol('δ')
g = Function('g')

And(
    Abs(t - x0) < delta,
    Abs(g(t) - x0) < a
)
```

```python
t1, t2, M = symbols('t1, t2, M')

lipschitz_condition = Abs(h(x, t2) - h(x, t1)) < M * Abs(t2 - t1)
lipschitz_condition
```

```python
x_2dot = Symbol('\ddot{x}')

Eq(x_2dot, f(x, t))
```

```python
v, v_dot, x_dot = symbols('v, \dot{v}, \dot{x}')

Eq(v_dot, f(x, t))
```
