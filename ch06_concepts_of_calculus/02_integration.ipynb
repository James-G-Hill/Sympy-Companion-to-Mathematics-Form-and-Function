 # 2. Integration

```python
from sympy import (
    Abs,
    And,
    Eq,
    Function,
    Idx,
    IndexedBase,
    integrate,
    Limit,
    Max,
    Min,
    Sum,
    Symbol,
    symbols
)
```

```python
a, b, x = symbols('a, b, x')
f = Function('f')

integrate(f(x), (x, a, b))
```

 There are 2 readings of the definite integral:

 1. the width *dx* of each strip is an infinitesimal increment in *x*; the integral represents an infinite sum of infiniteismal quantities.

 2. the desired area is approximated by a finite sum of the ares of the rectangles of finite width.

```python
A = IndexedBase('A')
x = IndexedBase('x')
i = symbols('i', cls=Idx)

min_i_f = Min(f(x[i] - x[i-1]))
max_i_f = Max(f(x[i] - x[i-1]))

And(
    min_i_f < A[i],
    A[i] < max_i_f
)
```

```python
L = Function('L')
sigma = Symbol('σ')
n = Symbol('n')

lower_sum = L(f(x), sigma).subs(f(x), Symbol('f'))
Eq(lower_sum, Sum(min_i_f, (i, 1, n)))
```

```python
U = Function('U')

upper_sum = U(f(x), sigma).subs(f(x), Symbol('f'))
Eq(upper_sum, Sum(max_i_f, (i, 1, n)))
```

```python
Abs(sigma)
```

```python
x = Symbol('x')

And(
    Eq(Limit(lower_sum, Abs(sigma), 0), integrate(f(x), (x, a, b))),
    Eq(Limit(upper_sum, Abs(sigma), 0), integrate(f(x), (x, a, b)))
)
```
