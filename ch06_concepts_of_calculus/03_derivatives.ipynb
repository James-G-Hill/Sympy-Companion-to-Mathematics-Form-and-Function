 # 3. Derivatives

```python
from spb import plot
from sympy import (
    diff,
    Function,
    IndexedBase,
    Limit,
    Symbol,
    symbols
)
```

```python
x = IndexedBase('x')
y = IndexedBase('y')

avg_rate_of_change = (y[2] - y[1]) / (x[2] - x[1])
avg_rate_of_change
```

```python
plot(
    avg_rate_of_change.subs({x[1]: 0, y[1]: 0, y[2]: 2}),
    (x[2], .5, 5)
)
```

```python
x = Symbol('x')

diff(x**2, x)
```

```python
dx = Symbol('dx')

plot(
    (((x + dx)**2 - x**2) / dx).subs(x, 1),
    (dx, 0, 1)
)
```

```python
dy, dt = symbols('dy, dt')

(dy / dx) * (dx / dt)
```

```python
f = Function('f')
a, h = symbols('a, h')

ratio = (f(x + h) - f(a)) / h
derivative = Limit(ratio, h, 0)
derivative
```

```python
class power_2(Function):
    @classmethod
    def eval(cls, n):
        return n**2

plot(
    ratio.subs({f: power_2, x: 1, a: x}),
    (h, 0, 1)
)
```
